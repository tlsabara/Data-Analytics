# Matplotlib

Mostrando praticas com Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Vari√°veis

dados_x = [round(random.random()*1000, 2) for i in range(15)]
dados_y = [ 30 + i ** random.random() / 5 for i in dados_x]
```

## modo Funcional do matplotlib

```python
# Uma plotagem simples mas com informativos.

plt.plot(dados_x)
plt.xlabel('Segundos')
plt.ylabel('Frequencia')

plt.title('Frequencia do Computador')
plt.show()
```

```python
# Uma plotagem com 2 graficos em uma figura

plt.subplot(1, 2, 1)
plt.plot(dados_x, 'r--')
plt.xlabel('Tempo')
plt.ylabel('Hz')
plt.title('Frequencia')

plt.subplot(1, 2, 2)
plt.plot(dados_y, 'b-*')
plt.xlabel('Tempo')
plt.ylabel('C*')
plt.title('Temperatura')
```
