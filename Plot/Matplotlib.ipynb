# Matplotlib

Mostrando praticas com Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Vari√°veis

dados_x = [round(random.random()*1000, 2) for i in range(15)]
dados_y = [ 30 + i ** random.random() / 5 for i in dados_x]
```

## modo Funcional do matplotlib

```python
# Uma plotagem simples mas com informativos.

plt.plot(dados_x)
plt.xlabel('Segundos')
plt.ylabel('Frequencia')

plt.title('Frequencia do Computador')
plt.show()
```

```python
# Uma plotagem com 2 graficos em uma figura

plt.subplot(1, 2, 1)
plt.plot(dados_x, 'r--')
plt.xlabel('Tempo')
plt.ylabel('Hz')
plt.title('Frequencia')

plt.subplot(1, 2, 2)
plt.plot(dados_y, 'b-*')
plt.xlabel('Tempo')
plt.ylabel('C*')
plt.title('Temperatura')
```

```python
# O jeito que eu mais gosto de trabalhar.

# Crio um objeto figure para o plot
fig = plt.figure()

# Crio um eixo base especificando a localiza√ß√£o dentro da figure
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# Realizo a plotagem dos dados e configura√ß√£o dos gr√°ficos.
axis1.plot(dados_y, 'r-*')
axis1.set_xlabel('Segundos')
axis1.set_ylabel('Temperatura')

axis2 = fig.add_axes([0.2, 0.65, 0.2, 0.2])
# Realizo a plotagem dos dados e configura√ß√£o dos gr√°ficos.
axis2.plot(dados_x )
axis2.set_xlabel('Segundos')
axis2.set_ylabel('Frenquencia')
```

```python
fig, axes = plt.subplots(figsize=(8,8))

axes.plot(dados_y, 'b', label='Core i7 12700k', linestyle='--')
axes.plot(dados_x, 'r-', label='Ryzen 7 7700x', marker='o')
axes.plot([x * 1.2 for x in dados_x],
          lw=20,
          alpha=0.1,
          color='#33dd44',
          label='Ryzen 7 7700x (Overclocked)')
axes.set_title('AMD x INTEL')
axes.set_ylabel('Pontua√ß√£o Single Core')
axes.set_xlabel('GB Used by Test')
axes.legend()
plt.show()
```
