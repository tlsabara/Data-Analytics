```python
# Exportando o Json
# Vamos Exportar 01 json, por cliente
# Apenas para dificultar


class Exporter:
    """
    Esta classe apenas exporta os arqivos
    """
    def __init__(self, dfex:pd.DataFrame):
        self.df = dfex
        self.destination = DEST_FOLDER / 'json_dmps'
        self.destination.mkdir(exist_ok=True)
        self.line_dict = None

    def controller_exporter(self):
        for i in self.df.values:
            self.__dict_mount(i)
            self.__export()


    def __export(self):
        json_mounted = json.dumps(self.line_dict, indent=4, cls=MyEncoder)
        with open(self.destination / f'client_{self.line_dict["ID"]}.json', 'w+') as file:
            file.writelines(json_mounted)


    def __dict_mount(self, line):
        self.line_dict = {}
        for b, val in enumerate(line):
            key = self.df.columns[b]
            self.line_dict[key] = val

class MyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(MyEncoder, self).default(obj)

```

```python
# Aqui realizei a exportação utilizando as classes criadas acima.
df_exporter = Exporter(df_to_split[to_json_columns])
df_exporter.controller_exporter()
```
