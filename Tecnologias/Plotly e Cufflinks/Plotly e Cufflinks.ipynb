```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns='alfa beta gama delta'.split())
```

```python
df.head(10)
```

```python
df.iplot(kind='scatter', x='alfa', y='delta', mode='markers')
```

```python
df.iplot('box')
```

```python
df2 = df.drop('delta', axis=1)
df2.gama = df.alfa.apply(lambda x: abs(x)**2)
df2.alfa = df.alfa.apply(lambda x: abs(x))
df2.beta = df.beta.apply(lambda x: abs(x))

```

```python
df2.iplot(kind='surface')
```

```python
df2.iplot(kind='bubble', x='alfa', y='beta', size='gama')
```

```python
df2.scatter_matrix(bins=50)
```
