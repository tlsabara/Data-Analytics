```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
df.head(10)
```

```python
grd = sns.PairGrid(df)
# grd.map(plt.scatter)
```

```python
grd.map_diag(plt.hist)
grd.map_upper(plt.scatter)
grd.map_lower(sns.kdeplot)
grd.map_lower(plt.scatter)
plt.show()
```

```python
grd2 = sns.FacetGrid(df, col='time', row='smoker')
grd2.map(plt.scatter, 'tip', 'total_bill')
grd2.map(sns.kdeplot, 'tip', 'total_bill', alpha=0.2)
```
